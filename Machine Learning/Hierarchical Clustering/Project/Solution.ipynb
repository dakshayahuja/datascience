### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Importing Data

```python
data = pd.read_csv("google_review_ratings.csv")
```

### Data Manipulation

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop('Unnamed: 25', axis=1)
data.columns = ['user_id', 'churches', 'resorts', 'beaches', 'parks',
                'theatres', 'museums', 'malls', 'zoo', 'restaurants',
                'pubs_bars', 'local_services', 'burger_pizza_shops',
                'hotels_other_lodgings', 'juice_bars', 'art_galleries',
                'dance_clubs', 'swimming_pools', 'gyms', 'bakeries',
                'beauty_spas', 'cafes', 'view_points', 'monuments', 'gardens']
```

```python
data['user_id'] = data['user_id'].str.split(' ',n=2, expand=True)[1]
data[data.columns] = data[data.columns].apply(pd.to_numeric, errors='coerce')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.fillna(data.mean())
data.isnull().sum()
```

### Exploratory Data Analysis

```python
fig = data.iloc[:,1:].hist(figsize=(20,20))
```

##### Visualising number of reviews for each category

```python
reviews = data[data.columns[1:]].astype(bool).sum().sort_values()
# print(reviews)
fig = plt.figure(figsize=(9,7))
plt.barh(np.arange(len(data.columns[1:])), reviews.values, align='center', alpha=0.7, color='green')
plt.yticks(np.arange(len(data.columns[1:])), reviews.index)
plt.title('Number of Reviews for each category')
plt.ylabel('Categories')
plt.xlabel('Number of reviews')
plt.show()
```

##### Visualising how many users gave reviews to categories

```python
no_reviews = data[data.columns[1:]].astype(bool).sum(axis=1).value_counts()
# print(no_reviews)
plt.bar(np.arange(len(no_reviews)), no_reviews.values, align='center', alpha=0.7, color='orange')
plt.xticks(np.arange(len(no_reviews)), no_reviews.index)
plt.title('Reviews vs Categories')
plt.ylabel('Number of reviews')
plt.xlabel('Number of categories')
plt.show()
```

##### Visualising average ratings per category

```python
averageRating = data[data.columns[1:]].mean().sort_values()
# print(averageRating)
fig = plt.figure(figsize=(9,7))
plt.barh(np.arange(len(data.columns[1:])), averageRating.values, align='center', alpha=0.7, color='steelblue')
plt.yticks(np.arange(len(data.columns[1:])), averageRating.index)
plt.title('Average Rating per Category')
plt.ylabel('Categories')
plt.xlabel('Average Rating')
plt.show()
```

### Finding the number of clusters using Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage
z = linkage(data, method='ward')
dendrogram = dendrogram(z)
plt.tick_params(labelbottom=False)
plt.title('Hierarchical Clustering')
```

#### Observing the dendrogram and seeing the problem statement we will go with 2 clusters i.e. positive and negative

### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2)
cluster = hc.fit_predict(data)
```

```python
data['cluster'] = cluster
data['cluster'] = data.cluster.apply(lambda x: "Positive" if x==1 else "Negative")
data.insert(1, 'cluster', data.pop('cluster'))
data
```
