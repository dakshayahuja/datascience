```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## Gradient Boosting

```python
x,y = make_classification(n_samples=1000, n_features=20, random_state=0)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
gbm = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
gbm.fit(x_train, y_train)
pred_1 = gbm.predict(x_test)
accuracy = accuracy_score(y_test, pred_1)
print(f'Accuracy is {accuracy*100}%')
```

## XGBoosting

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
xgb.fit(x_train, y_train)
pred_2 = xgb.predict(x_test)
accuracy = accuracy_score(y_test, pred_2)
print(f'Accuracy is {accuracy*100}%')
```

## AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
adaboost = AdaBoostClassifier(n_estimators=50, learning_rate=1, random_state=0)
adaboost.fit(x_train, y_train)
pred_4 = adaboost.predict(x_test)
accuracy = accuracy_score(y_test, pred_4)
print(f'Accuracy is {accuracy*100}%')
```

## CatBoost

```python
from catboost import CatBoostClassifier
catboost = CatBoostClassifier(iterations=100, learning_rate=0.1, max_depth=3, random_state=0)
catboost.fit(x_train, y_train)
pred_5 = catboost.predict(x_test)
accuracy = accuracy_score(y_test, pred_5)
print(f'Accuracy is {accuracy*100}%')
```
