Consider the age of 15 students enrolled in data science program.

age = [22, 29, 24, 32, 37, 26, 25, 45, 28, 22, 39, 47, 23, 36, 43]

**Hint:** Use the `skew()` from scipy library.

```python
def get_skewness_type(data):
    skewness = stats.skew(data)
    if skewness < -1:
        return "Highly Negative Skewed"
    elif skewness >= -1 and skewness < -0.5:
        return "Moderately Negative Skewed"
    elif skewness >= -0.5 and skewness <= 0.5:
        return "Symmetric"
    elif skewness > 0.5 and skewness <= 1:
        return "Moderately Positive Skewed"
    elif skewness > 1:
        return "Highly Positive Skewed"
    else:
        return "Unknown"
age = [22, 29, 24, 32, 37, 26, 25, 45, 28, 22, 39, 47, 23, 36, 43]
skewness_type = get_skewness_type(age)
print("Skewness Type:", skewness_type)
```

```python
def coefficient_of_variation(data):
    mean = np.mean(data)
    sd = np.std(data, ddof=1)
    cv = (sd / mean) * 100
    return cv

# Data for each blogger
sheldon_subscribers = np.array([3, 5.1, 2.9, 3.7, 4.2, 4.2, 4.3, 5.2, 3.7, 4.4, 5.2, 3.4])
ross_subscribers = np.array([4, 4.6, 3.1, 4.2, 5.4, 3.4, 5, 6.2, 4.2, 3.5, 4.1, 5.5])
joey_subscribers = np.array([4.3, 4, 4.5, 5.1, 6, 5, 4.8, 5.4, 5.1, 6.4, 4.2, 3.4])

# Calculate CV for each blogger
sheldon_cv = coefficient_of_variation(sheldon_subscribers)
ross_cv = coefficient_of_variation(ross_subscribers)
joey_cv = coefficient_of_variation(joey_subscribers)

print("Sheldon's Coefficient of Variation:", round(sheldon_cv, 2))
print("Ross's Coefficient of Variation:", round(ross_cv, 2))
print("Joey's Coefficient of Variation:", round(joey_cv, 2))

# Determine the most consistent blogger
most_consistent_blogger = min(
    ('Sheldon', sheldon_cv),
    ('Ross', ross_cv),
    ('Joey', joey_cv),
    key=lambda x: x[1]
)

print("The most consistent blogger is", most_consistent_blogger[0], "with a CV of", round(most_consistent_blogger[1], 2))
```
